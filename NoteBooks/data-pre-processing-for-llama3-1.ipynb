```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_ollama.llms import OllamaLLM
from langchain.text_splitter import RecursiveCharacterTextSplitter
import pandas as pd
import json
from tqdm import tqdm

```

```python
csv_file_path = "/kaggle/input/daily-star-articles-050824-050924/articles_data.csv"
json_file_path = "/kaggle/working/hasina.json"

json_array = []
```

```python
template = """Task:Assist the user in creating questions by using the input. You will just return the question
and the question will be just one line.
input: {input}
"""

text_splitter = RecursiveCharacterTextSplitter(
    separators=[".", "\n\n", "\n\n\n", "\n\n\n\n", "\n\n\n\n\n"],
    chunk_size=600,
    chunk_overlap=70,
    length_function=len,
    is_separator_regex=False,
)

def add_to_json_array(data):
    json_array.append(data)
```

```python
prompt = ChatPromptTemplate.from_template(template)

model = OllamaLLM(model="smollm")

chain = prompt | model
```

```python
def main():
    df = pd.read_csv(csv_file_path)
    chain = prompt | model
    for index, row in tqdm(df.iterrows(), desc="Main articles", position=0):
        article = text_splitter.split_text(row["body"])

        for i, chunk in tqdm(
            enumerate(article), desc="  Chunk of articles", position=1
        ):
            question = chain.invoke({"input": chunk})

            if i == 0:
                add_to_json_array(
                    {"instruction": question, "input": row["headline"], "output": chunk}
                )
            else:
                add_to_json_array(
                    {"instruction": question, "input": "", "output": chunk}
                )

    with open(json_file_path, "w") as json_file:
        json.dump(json_array, json_file, indent=2)
```

```python
main()
```
